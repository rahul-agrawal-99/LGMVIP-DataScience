### <font color='blue'>Image to Pencil Sketch with Python</font>

####   Required Libraries 

```python

import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

```python
# We need to read the image in RBG format and then convert it to a grayscale image. This will turn an image into a
# classic black and white photo. Then the next thing to do is invert the grayscale image also called negative image, 
# this will be our inverted grayscale image. Inversion can be used to enhance details. Then we can finally create the
# pencil sketch by mixing the grayscale image with the inverted blurry image. This can be done by dividing the grayscale 
# image by the inverted blurry image. Since images are just arrays, we can easily do this programmatically using the divide 
# function from the cv2 library in Python.
```

```python
#  simple function to display image

def show(img , title="default_Title" , fig_size=(7,7)):
    plt.figure(figsize=fig_size)
    plt.imshow(cv.cvtColor(img,cv.COLOR_BGR2RGB))  
    plt.title(title)
    plt.axis('off')
    plt.plot()


def img_compare(img1 , img2 , title1= "First Image", title2= "second Image"):
    fig = plt.figure(figsize=(20, 20))
    fig.add_subplot(2,2, 1)
    plt.imshow(cv.cvtColor(img1,cv.COLOR_BGR2RGB))
    plt.axis('off')
    plt.title(f"{title1}                              VS")
    fig.add_subplot(2,2, 2)
    plt.imshow(cv.cvtColor(img2,cv.COLOR_BGR2RGB))
    plt.axis('off')
    plt.title(title2)
```

```python
img = cv.imread('img.jfif')

show(img , "main Image")

```

####  <font color='gray'>Conveting to GrayScale Image</font>

```python
img_gray = cv.cvtColor(img,cv.COLOR_RGB2GRAY)

show(img_gray ,"Gray Scale image")
```

```python
img_invert = cv.bitwise_not(img_gray)

show(img_invert)
```

```python
blur= cv.GaussianBlur(img_invert, (17,21),sigmaX=0, sigmaY=0)

show(blur,"blur image")
```

```python
img_compare(blur,img_gray , " Blur Image" ,"Gray Image")
```

```python
a=cv.divide(img_gray,250-blur, scale=240)


show(a , "pencil Sketch")

```

```python
fig = plt.figure(figsize=(20, 20))

fig.add_subplot(2,2, 1)
  
plt.imshow(cv.cvtColor(a,cv.COLOR_BGR2RGB))
plt.axis('off')
plt.title("First")
  
fig.add_subplot(2,2, 2)
  
plt.imshow(cv.cvtColor(img,cv.COLOR_BGR2RGB))
plt.axis('off')
plt.title("Second")
```
